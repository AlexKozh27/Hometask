```python
root = tree.getroot()
tree
```

```python
def pagerank(M, num_iterations = 100, d = 0.85):
    N = M.shape[1]
    x = np.random.rand(N, 1)
    x = x / np.linalg.norm(x, 1)
    M_hat = (d * M + (1 - d) / N)
    for i in range(num_iterations):
        x = M_hat @ x
    return x / x.sum()
```

```python
def printing(name, data):
    print(name)
    display(data)
```

```python
names = []
for i in tree.findall('MetaNetwork/nodes/nodeclass/node'):
    names.append(i.attrib.get('id'))
names

np.random.seed(5)

data_padgm = pd.DataFrame(np.random.randn(16, 16), index=names, columns=names)
links_padgm = tree.findall("MetaNetwork/networks/network[@id = 'PADGM']")[0].findall('link')
for link in links_padgm:
    data = link.attrib
    data_padgm.loc[data.get('source'), data.get('target')] = data.get('value')

data_padgb = pd.DataFrame(np.random.randn(16, 16), index=names, columns=names)
links_padgb = tree.findall("MetaNetwork/networks/network[@id = 'PADGB']")[0].findall('link')
for link in links_padgb:
    data = link.attrib
    data_padgb.loc[data.get('source'), data.get('target')] = data.get('value')
```

```python
printing('PADGM', data_padgm)
```

```python
printing('PADGB', data_padgb)
```

```python
padgm_arr = data_padgm.to_numpy(dtype='float')
div = padgm_arr.sum(axis=1)
for i in range(len(div)):
    if div[i] == 0:
        div[i] += 1
pr_padgm = pagerank(padgm_arr / div)
prm_padgm = np.reshape(pr_padgm, (1, 16))[0]
```

```python
padgb_arr = data_padgb.to_numpy(dtype='float')
div = padgb_arr.sum(axis=1) 
for i in range(len(div)):
    if div[i] == 0:
        div[i] += 1
pr_padgb = pagerank(padgb_arr / div)
pr_padgb = np.reshape(pr_padgb, (1, 16))[0]
```

```python
relations = pd.DataFrame({"Marriage" : prm_padgm, "Business" : pr_padgb, "Total weighted" : (prm_padgm + pr_padgb) / 2}, index=names)
relation.sort_values(by="Total weighted", ascending=False)
```
